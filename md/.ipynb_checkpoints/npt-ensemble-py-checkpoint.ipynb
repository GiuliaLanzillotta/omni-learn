```python
psel = traj.sigma == 3.4e-10
nsel = np.logical_not(psel)
psel.sum()
```

```python
plt.figure()

for r, l, temp in zip(traj.r, traj.L, traj.T):
#     print(r.shape)
#     break
    plt.cla()
    #plt.xlim(-lim/2, lim/2)
    #plt.ylim(-lim/2, lim/2)
    plt.scatter(*r[psel].T, c='r')
    plt.scatter(*r[nsel].T, c='b')
    plt.plot([-l/2, -l/2, l/2, l/2, -l/2], [-l/2, l/2, l/2, -l/2, -l/2], ls='--')
    plt.xlabel(str(temp))
    plt.pause(0.033)
```
