```julia
using DifferentialEquations, PyPlot, SpecialFunctions, DataStructures, HDF5 #BenchmarkTools
#gr();

include("Lab.jl");
include("Sim.jl");
include("Forces.jl");

grid(w,h) = hcat([collect(i) for i in Iterators.product(w, h)]...);
```

```julia
N = 127
atoms = [Ytterbium for i=1:N];
```

```julia
config = SimInfo(DuanTrap, atoms, N, [atom.mass for atom in atoms], [atom.charge for atom in atoms]);
```

```julia
# init conditions
#z₀ = zeros(3,N,2)
r₀ = randn(3, N) * 1e-5
```

```julia
function dynamics(r, p, t)
    ṙ = zeros(3, config.N)

    Coulomb!(ṙ, r, t, config)
    #Harmonic!(f, r, t, config, ω)
    Paul_E_Field!(ṙ, r, t, config)
    
    for i=1:config.N # force to acceleration
        ṙ[:,i] ./= config.m[i]
    end
    return ṙ * p
end;
```

```julia
nframes = 1000
stop = 1e-3
#tspan = (0., frames*timestep)
tspan = (0., stop)
lr = 1e-11
prob = ODEProblem(dynamics, r₀, tspan, lr);
stop / nframes
```

```julia
r = sol.u[end]
tspan = (2e2*stop, 4e2*stop)
prob = ODEProblem(dynamics, r, tspan);
tspan
```

```julia
sol = solve(prob);
println("Done")#  dense=false, adaptive=false, dt=timestep, saveat=0:timestep:tspan[2])
```

```julia
# init conditions
γ = 1e-20
z₀ = randn(3,N,2) * 1e-6
z₀[1:end-1,:,2] = grid(-1:2/11:1, -1:2/11:1)[:,1:127] * 2e-5
z₀[:,:,1] *= 0
z₀
```

```julia
function dynamics(z, p, t)
    ż = zeros(3, config.N, 2)
    f = view(ż, :, :, 1) # force
    
    ṙ = view(z, :, :, 1)
    r = view(z, :, :, 2)

    Damping!(f, ṙ, t, config, γ)
    #Random_Heating!(f, ṙ, t, config)
    #Laser_Cooling!(ṙ, t, config)
    #Scattering_Force!(f, ṙ, t, config, laser_settings)
    
    Coulomb!(f, r, t, config)
    #Harmonic!(f, r, t, config, ω)
    Paul_E_Field!(f, r, t, config)
    
    for i=1:config.N # force to acceleration
        f[:,i] ./= config.m[i] / 1e-3
    end
    ż[:,:,2] = ṙ
    return ż
end;
```

```julia
dynamics(z₀,[],1e-8)
```

```julia
start = 0
stop = 1e-5
#tspan = (0., frames*timestep)
tspan = (start, stop)
prob = ODEProblem(dynamics, z₀, tspan);
tspan
```

```julia
solve(prob)
print("Done")
```
