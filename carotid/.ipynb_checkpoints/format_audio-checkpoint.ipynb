```python
import sys, os
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import pydub
import torch
```

```python
def crawl(d, cond):
    if os.path.isdir(d):
        options = []
        for f in os.listdir(d):
            path = os.path.join(d, f)
            options.extend(crawl(path, cond))
        return options
    if cond(d):
        return [d]
    return []
```

```python
audio_files = crawl('/home/fleeb/workspace/ml_datasets/audio/yt/test', lambda x: '.mp3' in x)
print(len(audio_files))
for n in audio_files:
    print(n)
```
